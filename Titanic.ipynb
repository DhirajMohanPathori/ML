```python
import numpy as np
import pandas as pd
```

# Read Data

```python
train_data = pd.read_csv ("Titanic/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv ("Titanic/test.csv")
test_data.head()
```

# Data Cleaning

```python
train_data.info()
```

```python
train_data.drop(['Name', 'Fare'], axis = 1, inplace = True)
train_data.drop('Cabin', axis = 1, inplace = True)
train_data.head()
```

```python
test_data.drop(['Fare'], axis = 1, inplace = True)
test_data.drop('Cabin', axis = 1, inplace = True)
test_data.head()
```

```python
numeric_cols = train_data.select_dtypes (include = np.number).columns.tolist()
```

```python
categorical_cols = train_data.select_dtypes (include = 'object').columns.tolist()
```

```python
numeric_cols_2 = test_data.select_dtypes (include = np.number).columns.tolist()
```

```python
categorical_cols_2 = test_data.select_dtypes (include = "object").columns.tolist()
```

## Missing Data

```python
train_data.fillna(train_data['Age'].mean(), inplace = True)
```

```python
test_data.fillna(test_data['Age'].mean(), inplace = True)
```

```python
train_data[train_data['Embarked'] == 'NaN']
```

## One Hot Encoding

```python
low_cardinality_cols = [col for col in categorical_cols if train_data[col].nunique() < 10]
```

```python
high_cardinality_cols = list(set(categorical_cols)-set(low_cardinality_cols))
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols_train = pd.get_dummies(train_data[low_cardinality_cols])
OH_cols_test = pd.get_dummies(test_data[low_cardinality_cols])
```

```python
OH_cols_train.index = train_data.index
OH_cols_test.index = test_data.index
```

```python
num_train_data = train_data.drop(categorical_cols, axis=1)
num_test_data = test_data.drop(categorical_cols_2, axis=1)
```

```python
OH_train_data = pd.concat([num_train_data, OH_cols_train], axis=1)
OH_test_data = pd.concat([num_test_data, OH_cols_test], axis=1)
```

```python
train_target = OH_train_data['Survived']
```

```python
train_feature = OH_train_data.drop('Survived', axis = 1)
```

```python
train_feature.drop('Embarked_29.69911764705882', axis=1, inplace=True)
```

# Model Training

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()

model.fit(train_feature, train_target)
predictions = model.predict(OH_test_data)
```

```python
predictions = predictions.astype(np.int32)
```
